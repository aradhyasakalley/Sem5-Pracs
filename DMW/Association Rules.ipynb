Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
```

```python
df = pd.read_csv('Groceries_dataset.csv')
```

```python
df.head()
```

```python
min_support = float(input("Enter the minimum support: "))
confidence = float(input("Enter the minimum confidence: "))
```

```python
df_processed = df.applymap(lambda x: 1 if x != 0 else 0)
```

```python
df.head()
```

```python
frequent_itemsets = apriori(df_processed, min_support=min_support, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=confidence)
```

```python
print("\nAssociation Rules:")
print(rules)
```

FPTree

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
import pandas as pd
```

```python
transactions = [
    ({'f', 'a', 'c', 'd', 'g', 'i', 'm', 'p'}, 100),
    ({'a', 'b', 'c', 'f', 'l', 'm', 'o'}, 200),
    ({'b', 'f', 'h', 'j', 'o', 'w'}, 300),
    ({'b', 'c', 'k', 's', 'p'}, 400),
    ({'a', 'f', 'c', 'e', 'l', 'p', 'm', 'n'}, 500)
]
```

```python
print(transactions)
```

```python
transactions_list = [list(t[0]) for t in transactions]
print(transactions_list)
```

```python
te = TransactionEncoder()
te_ary = te.fit(transactions_list).transform(transactions_list)
df_custom = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
print(df_custom)
```

```python
min_support = 0.2  # Adjust the min_support value as needed
frequent_itemsets_custom = fpgrowth(df_custom, min_support=min_support, use_colnames=True)
```

```python
print("\nFrequent Patterns (FP Growth) - Custom Dataset:")
for index, row in frequent_itemsets_custom.iterrows():
    tids = [tid for tid, transaction in zip([t[1] for t in transactions], transactions_list) if set(row['itemsets']).issubset(set(transaction))]
    print(f"TID(s): {', '.join(map(str, tids))} - Pattern: {set(row['itemsets'])}")
```
