```python
a = [1,2,3,4,5]
myvar = pd.Series(a)
print(myvar)
```

```python
a = [1,2,3,4,5]
myvar = pd.Series(a,index=['x','y','z','a','b'])
print(myvar)
```

```python
myvar = pd.Series({'day1':'monday','day2':'tuesday'})
print(myvar)
```

```python
data = {
    'days' : [1,2,3,4,5],
    'calories' :[20,10,30,40,50]
}
myvar = pd.DataFrame(data)
print(myvar)
print(myvar.loc[0])
print(myvar.loc[[0,1]])
```

```python
data = pd.read_csv('card.csv')
```

```python
data.head()
```

```python
new_data = data.dropna()
print(new_data.to_string())
```

```python
new_data = data.dropna(subset=['Period'])
print(new_data.to_string())
```

```python
x = data['Period'].mean()
print(x)
```

```python
data['Period'].fillna(x,inplace=True)
```

```python
data['Period']
```

```python
x = data['Period'].median()
print(x)
print(data['Period'])
```
